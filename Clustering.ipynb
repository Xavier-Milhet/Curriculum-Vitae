```
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
```

```
data = datasets.load_digits()
X, y = data.data[:500], data.target[:500]
```

```
sc = StandardScaler()
X_scale = sc.fit_transform(X)
```

```
pca = PCA(n_components=32)
pca.fit(X_scale)
X_pca = pca.transform(X_scale)

tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X_pca)
```

```
plt.figure()
plt.scatter(X_tsne[:,0],X_tsne[:,1],c=y)
plt.show()
```

```
link = 'ward'
linked = linkage(X_tsne, link)
dendrogram(linked)
plt.show()
```

```
ac = AgglomerativeClustering(n_clusters=10,linkage=link)
ac.fit(X_tsne)

_ = plt.figure()
plt.scatter(X_tsne[:,0],X_tsne[:,1],c=ac.labels_)
plt.show()
```

```
labels = [list(y[ac.labels_==x]) for x in set(ac.labels_)]
cluster_labels = {y:[max(x,key=x.count) for x in labels][y] for y in range(10)}
classif_labels = [cluster_labels[x] for x in ac.labels_]
```

```
print(f'Classification score: %.03f'%(sum(classif_labels==y)/X.shape[0]))
```
